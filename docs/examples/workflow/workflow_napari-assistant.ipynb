# Generating workflows with napari

This page describes how to generate custom workflows using napari using the calculator-like interface from [`napari-assistant`](https://github.com/haesleinhuepf/napari-assistant). In many ways, this should be reminiscent of using the `Macro Recorder` in FIJI/ImageJ, but you will hopefully find it has more flexibility and advantages. Ultimately, our goal is to easily reproduce image processing steps with a [`napari-workflows`](https://github.com/haesleinhuepf/napari-workflows) `.yaml` file and utilize it for batch processing with `napari-ndev:Workflow Widget`. 

### napari-assistant

In order to generate workflows with napari, you may like to `pip install napari-ndev[workflow]` to install the napari-assistant and other cooperating plugins. You can then access the assistant via the `Plugins menu -> Assistant (clesperanto)`

 A thorough tutorial on how to use napari-assistant, including video, [can be found here](https://haesleinhuepf.github.io/BioImageAnalysisNotebooks/70_workflows/napari-assistant.html). The assistant is quite flexible and both functions and parameters are modifiable on-the-fly, but can be overall quirky at times. 

```python
import napari
from napari.utils import nbscreenshot

viewer = napari.Viewer()
viewer.window.resize(500, 800) # w x h
viewer.window.add_plugin_dock_widget('napari-assistant')
nbscreenshot(viewer)
```
